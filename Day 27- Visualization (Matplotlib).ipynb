#### Matplotlib Overview:
- **Purpose**: Matplotlib is a powerful library for creating static, animated, and interactive visualizations in Python.
- **Features**: Provides functions for creating various types of plots, including:
  - Line plots
  - Scatter plots
  - Bar plots
  - Histograms
  - Heatmaps
  - And more
- **Customization**: Highly customizable, allowing users to control aspects such as colors, line styles, markers, axis limits, legends, annotations, and more.
- **Compatibility**: Works seamlessly with NumPy arrays and Pandas DataFrames.
- **Integration**: Often used in conjunction with Jupyter Notebooks for interactive plotting and data exploration.

#### Basic Plotting Commands:
- **Line Plot**: `plt.plot(x, y)`
- **Scatter Plot**: `plt.scatter(x, y)`
- **Bar Plot**: `plt.bar(x, height)`
- **Histogram**: `plt.hist(data, bins)`
- **Box Plot**: `plt.boxplot(data)`

#### Customization Commands:
- **Labels & Title**:
  - X-axis Label: `plt.xlabel('label')`
  - Y-axis Label: `plt.ylabel('label')`
  - Title: `plt.title('title')`
- **Colors & Styles**:
  - Line Color: `color='color_name'`
  - Line Style: `linestyle='style_name'`
  - Marker Style: `marker='marker_name'`
- **Axis Limits**:
  - X-axis Limit: `plt.xlim(xmin, xmax)`
  - Y-axis Limit: `plt.ylim(ymin, ymax)`
- **Legends & Annotations**:
  - Legend: `plt.legend(['label1', 'label2'])`
  - Text Annotation: `plt.text(x, y, 'text')`

#### Displaying & Saving Plots:
- **Display**: `plt.show()`
- **Save Figure**: `plt.savefig('filename.png')`

</span>
    

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y , c = '#223BDC',alpha=.9,marker='X')
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my x vs y")
plt.grid()
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,'--r')
plt.xlabel("This is my x data")
plt.ylabel("This is y data")
plt.title("This is my sin wave graph")
plt.grid()
```

```python
x = ['a', 'b' , 'c','d','e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.bar(x,y,color = '#7E10D5')
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("this is my bar plot")
plt.grid()
```

```python
plt.figure(figsize=(5,3))
plt.barh(x,y,color = '#7E10D5')
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("this is my bar plot")
plt.grid()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z,c='#9327E8',alpha = .9)
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.show()
```
