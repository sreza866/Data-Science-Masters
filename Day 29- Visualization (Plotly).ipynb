### Plotly

<span style = 'font-size:0.8em;'>
    
Plotly is a versatile and powerful open-source graphing library that enables the creation of interactive, web-based visualizations. It supports a variety of chart types and is compatible with several programming languages, including Python, R, MATLAB, and JavaScript. Plotly is particularly popular in data science and analytics for its ability to produce publication-quality graphics and its integration with data manipulation tools.

Here are some key features and uses of Plotly:

### Key Features

1. **Interactive Graphs**: Plotly allows users to create interactive charts that can be embedded in web applications or Jupyter Notebooks. Users can zoom, pan, and hover over data points to get more detailed information.
  
2. **Wide Range of Chart Types**: Plotly supports a broad spectrum of chart types, including:
   - Line charts
   - Bar charts
   - Scatter plots
   - Heatmaps
   - Box plots
   - Histograms
   - Pie charts
   - 3D plots
   - Geographic maps

3. **Customizable**: Charts can be highly customized, from colors and labels to annotations and shapes. This flexibility allows for the creation of tailored visualizations that meet specific presentation needs.

4. **Integration**: Plotly integrates seamlessly with other libraries and tools, such as Pandas for data manipulation and Dash for creating interactive web applications. It also supports exporting to static image formats like PNG and SVG.

5. **Ease of Use**: The library is designed to be user-friendly with straightforward syntax, making it accessible for users with varying levels of programming experience.

#### Basic Imports
```python
import plotly.graph_objects as go  # Primary graphing library
import plotly.express as px        # Simplified interface for creating plots
```

### Creating Different Plot Types

#### Line Plot
```python
fig = go.Figure(go.Scatter(x=[1, 2, 3, 4], y=[10, 11, 12, 13], mode='lines'))
fig.show()
```

#### Scatter Plot
```python
fig = go.Figure(go.Scatter(x=[1, 2, 3, 4], y=[10, 11, 12, 13], mode='markers'))
fig.show()
```

#### Bar Chart
```python
fig = go.Figure(go.Bar(x=['A', 'B', 'C'], y=[10, 20, 30]))
fig.show()
```

#### Histogram
```python
fig = go.Figure(go.Histogram(x=[1, 1, 2, 2, 3, 3, 4, 4]))
fig.show()
```

#### Box Plot
```python
fig = go.Figure(go.Box(y=[1, 2, 3, 4, 5, 6, 7]))
fig.show()
```

#### Heatmap
```python
fig = go.Figure(go.Heatmap(z=[[1, 20, 30], [20, 1, 60], [30, 60, 1]]))
fig.show()
```

#### Pie Chart
```python
fig = go.Figure(go.Pie(labels=['A', 'B', 'C'], values=[10, 20, 30]))
fig.show()
```

#### 3D Scatter Plot
```python
fig = go.Figure(go.Scatter3d(x=[1, 2, 3], y=[1, 2, 3], z=[1, 2, 3], mode='markers'))
fig.show()
```

#### Geographic Map
```python
fig = go.Figure(go.Scattergeo(lat=[40.7128, 34.0522], lon=[-74.0060, -118.2437], mode='markers'))
fig.show()
```

### Customizing Plots

#### Adding Titles and Axis Labels
```python
fig.update_layout(
    title='Plot Title',
    xaxis_title='X Axis Title',
    yaxis_title='Y Axis Title'
)
```

#### Updating Marker Styles
```python
fig.update_traces(marker=dict(size=12, color='rgb(51,204,153)', symbol='circle'))
```

#### Adding Annotations
```python
fig.add_annotation(
    x=2, y=12,
    text="Important Point",
    showarrow=True,
    arrowhead=1
)
```

#### Setting Plot Size
```python
fig.update_layout(width=800, height=600)
```

### Using Plotly Express

#### Simple Line Plot
```python
fig = px.line(x=[1, 2, 3], y=[10, 20, 30], title='Simple Line Plot')
fig.show()
```

#### Simple Scatter Plot
```python
fig = px.scatter(x=[1, 2, 3], y=[10, 20, 30], title='Simple Scatter Plot')
fig.show()
```

#### Simple Bar Chart
```python
fig = px.bar(x=['A', 'B', 'C'], y=[10, 20, 30], title='Simple Bar Chart')
fig.show()
```

### Saving Plots
```python
fig.write_image("figure.png")  # Save as PNG
fig.write_html("figure.html")  # Save as HTML
```

````python
import seaborn as sns
tips = sns.load_dataset("tips")
````

````python
tips
````

````python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill, y = tips.tip, mode = 'markers'))
# Add a title and axis labels to the plot
fig.update_layout(
    title='Total Bill vs Tip',
    xaxis_title='Total Bill',
    yaxis_title='Tip'
)
fig.show("png")
````

````python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill, y = tips.tip, mode = 'lines'))
fig.update_layout(
    title='Total Bill vs Tip',
    xaxis_title='Total Bill',
    yaxis_title='Tip'
)
fig.show("png")
````

````python
fig = go.Figure()
fig.add_trace(go.Bar(x = [1,2,3,4,5,3], y = [3,4,5,5,6,7]))
fig.update_layout(title='x_list vs y_list')
fig.show("png")
````

````python
fig = go.Figure()
fig.add_trace(go.Histogram(x = tips["total_bill"]))
# Update layout with title and axis labels
fig.update_layout(
    title='Distribution of Total Bill',
    xaxis_title='Total Bill',
    yaxis_title='Frequency',
    xaxis=dict(
        title_font=dict(size=18, color='darkblue')
    ),
    yaxis=dict(
        title_font=dict(size=18, color='darkblue')
    )
)
fig.show("png")
````

````python
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill,y =tips.tip,mode = 'markers',marker_size = 10*tips['size']))
# Update layout with title and axis labels
fig.update_layout(
    title='Distribution of Total Bill vs Size',
    xaxis_title='Total Bill',
    yaxis_title='Tip',
    xaxis=dict(
        title_font=dict(size=18, color='darkblue')
    ),
    yaxis=dict(
        title_font=dict(size=18, color='darkblue')
    )
)
fig.show("png")
````
